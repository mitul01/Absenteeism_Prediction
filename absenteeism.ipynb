```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
raw_data=pd.read_csv('absenteeism.csv')
raw_data.head()
```

```python
raw_data1=raw_data.drop('ID',axis=1)
```

```python
raw_data1.head()
```

```python
raw_data1.describe(include='all')
```

```python
raw_data.isnull().sum()
```

```python
raw_data1['Reason for Absence'].unique()
```

```python
reason_absence=pd.get_dummies(raw_data1['Reason for Absence'],drop_first=True)
```

```python
reason_absence
```

```python
reason1=reason_absence.loc[:,1:14].max(axis=1)
reason2=reason_absence.loc[:,15:17].max(axis=1)
reason3=reason_absence.loc[:,18:21].max(axis=1)
reason4=reason_absence.loc[:,22:28].max(axis=1)
```

```python
data_reason_dummies=pd.concat([raw_data1,reason1,reason2,reason3,reason4],axis=1)
```

```python
data_reason_dummies.head()
```

```python
columns_updated=(['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours','Reason1', 'Reason2', 'Reason3','Reason4'])
data_reason_dummies.columns=columns_updated
```

```python
columns_reordered=(['Reason1', 'Reason2', 'Reason3','Reason4','Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours',])

data_reason_dummies=data_reason_dummies[columns_reordered]
data_reason_dummies.head()
```

```python
data_reason_dummies=data_reason_dummies.drop('Reason for Absence',axis=1)
```

```python
data_reason_dummies.head()
```

```python
data_reason_dummies['Date'] = pd.to_datetime(data_reason_dummies['Date'], format = '%d/%m/%Y')
```

```python
months=[]
for i in range(700):
    months.append(data_reason_dummies['Date'][i].month)
```

```python
data_months=data_reason_dummies.copy()
data_months['Month']=months
```

```python
weekday=[]
for i in range(700):
    weekday.append(data_reason_dummies['Date'][i].weekday())
```

```python
weekday
```

```python
data_weekdays=data_months.copy()
data_weekdays['weekday']=weekday
```

```python
data_weekdays
```

```python
data1=data_weekdays.drop('Date',axis=1)
```

```python
data1.head()
```

```python
data1['Education']=data1['Education'].map({1:0,2:1,3:1,4:1})
```

```python
data1
```

```python
data1.columns.values
```

```python
new_columns=(['Reason1', 'Reason2', 'Reason3', 'Reason4','Month',
       'weekday','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours'])
data1=data1[new_columns]
```

```python
data1.head()
```

```python
data_preprocessed=data1.copy()
```

```python
data_preprocessed.to_csv('Absenteeism_preprocessed_data.csv',index=False)
```
